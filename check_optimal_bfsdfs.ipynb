```python

from collections import deque

Graph = {
    "A": ["B", "C"],
    "B": ["D", "E"],
    "C": ["F"],
    "D": [],
    "E": [],
    "F": []
}

def bfs(graph, start, target):
    color = {i: "White" for i in graph}
    distance = {i: float("inf") for i in graph}
    parent = {i: None for i in graph}
    Q = deque()

    color[start] = "Grey"
    distance[start] = 0
    Q.append(start)

    while Q:
        i = Q.popleft()
        for u in graph[i]:
            if u == target:
                distance[u] = distance[i] + 1
                print(f"[BFS] Target '{target}' found at distance: {distance[u]}")
                return
            if color[u] == "White":
                color[u] = "Grey"
                distance[u] = distance[i] + 1
                parent[u] = i
                Q.append(u)
        color[i] = "Black"

    print("[BFS] Target not found.")


found = False 

def dfs(graph, start, target):
    global time, found
    color = {i: "White" for i in graph}
    parent = {i: None for i in graph}
    discovery = {}
    finish = {}
    time = 0

    def dfs_visit(u):
        global time, found
        if found:
            return

        color[u] = "Grey"
        time += 1
        discovery[u] = time

        for v in graph[u]:
            if found:
                return
            if v == target:
                time += 1
                discovery[v] = time
                print(f"[DFS] Target '{target}' found after {time} steps.")
                found = True
                return
            if color[v] == "White":
                parent[v] = u
                dfs_visit(v)

        color[u] = "Black"
        time += 1
        finish[u] = time

    dfs_visit(start)

    if not found:
        print("[DFS] Target not found.")

print("=== Running BFS ===")
bfs(Graph, "A", "B")

print("\n=== Running DFS ===")
found = False
dfs(Graph, "A", "E")
```
